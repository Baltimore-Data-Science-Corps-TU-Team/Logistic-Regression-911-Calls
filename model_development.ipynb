```python
#X = data.iloc[:,0:363]
X = df2

#y — what is the severity of the crime? (binary: “1”, means “Violent”, “0” means “Nonviolent”)
#y = data.iloc[:,363]
y = data["Severity"]
```

### Test + Train Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
#X_train, X_test, y_train, y_test = train_test_split(data.drop('Severity',axis=1), data['Severity'], test_size=0.30, random_state=101)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Fit only to the training data
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Model Prediction

```python
from sklearn import linear_model
logreg = linear_model.LogisticRegression(C=1e5)
```

```python
logreg.fit(X_train, y_train)
logreg_pred = logreg.predict(X_test)
logreg
```

```python
from __future__ import division
(sum(y_test==logreg_pred))/len(logreg_pred)
```

```python
from sklearn.metrics import confusion_matrix
logreg_tab = confusion_matrix(y_test, logreg_pred)
logreg_tab
```

```python
sns.heatmap(pd.DataFrame(confusion_matrix(y_test,logreg_pred)))
plt.show()
```

The result is telling us that we have 60292+0 correct predictions and 0+68785 incorrect predictions.

```python
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
print('Accuracy of logistic regression classifier on train set: {:.2f}'.format(logreg.score(X_train, y_train)))
```

```python
from sklearn import metrics
print (metrics.classification_report(y_test, logreg_pred))
```
